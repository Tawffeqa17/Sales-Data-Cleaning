## DATA CLEANING

```python
sd.isnull().sum() 
```

```python
Order=0
Product=0            
City=0             
Quantity=0         
Price=0      
Total=0
s=['UNKNOWN','ERROR']
for i in sd.index:
    if sd.loc[i,'Order_ID'] in s:
      Order+=1
    if sd.loc[i,'Product'] in s:
      Product+=1
    if sd.loc[i,'City'] in s:
      City+=1
    if sd.loc[i,'Quantity'] in s:
      Quantity+=1
    if sd.loc[i,'Unit_Price'] in s:
      Price+=1
    if sd.loc[i,'Total_Amount'] in s:
      Total+=1
print(f'''Count of UNKNOWN ~ 
order    ={Order}
product  ={Product}
city     ={City}
quantity ={Quantity}
price    ={Price}
total    ={Total}''')
```

```python
mode_product = sd["Product"].mode()[0]

for i in sd.index:
    if sd.loc[i, "Product"] == "UNKNOWN":
        sd.loc[i, "Product"] = mode_product

print("Done")
```

```python
(sd["Product"] == "UNKNOWN").sum()
```

#### After replaced

```python
Order=0
Product=0            
City=0             
Quantity=0         
Price=0      
Total=0
s=['UNKNOWN','ERROR']
for i in sd.index:
    if sd.loc[i,'Order_ID'] in s:
      Order+=1
    if sd.loc[i,'Product'] in s:
      Product+=1
    if sd.loc[i,'City'] in s:
      City+=1
    if sd.loc[i,'Quantity'] in s:
      Quantity+=1
    if sd.loc[i,'Unit_Price'] in s:
      Price+=1
    if sd.loc[i,'Total_Amount'] in s:
      Total+=1
print(f'''Count of UNKNOWN ~ 
order    ={Order}
product  ={Product}
city     ={City}
quantity ={Quantity}
price    ={Price}
total    ={Total}''')
```

#### Replacing temp values

```python
for i in sd.index:
  if sd.loc[i,'Quantity'] in s:
       sd.loc[i,'Quantity']='0'
sd.fillna({'Quantity':'0'},inplace=True)
print('Done')
```

```python
for i in sd.index:
  if sd.loc[i,'Unit_Price'] in s:
       sd.loc[i,'Unit_Price']='0'
sd.fillna({'Unit_Price':'0'},inplace=True)
print('Done')
```

```python
for i in sd.index:
  if sd.loc[i,'Total_Amount'] in s:
       sd.loc[i,'Total_Amount']='0'
sd.fillna({'Total_Amount':'0'},inplace=True)
print('Done')
```
